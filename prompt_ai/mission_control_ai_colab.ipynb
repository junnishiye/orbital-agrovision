```python
from getpass import getpass
from openai import OpenAI

API_KEY = getpass("Digite sua chave de API: ")

client = OpenAI(
    api_key=API_KEY
)
```

```python
"dados disponiveis no repositorio(que seriam usados sem necessidade de recolocar aqui), porem, por estar no colab recolocamos aqui"
system_prompt = """
Você é o Mission Control AI da missão TerraGuard, parte do projeto Orbital AgroVision.

A Orbital AgroVision é um sistema inteligente de monitoramento espacial aplicado ao agronegócio sustentável. A missão TerraGuard simula o uso de sensores orbitais, módulos espaciais e análise computacional para monitorar áreas agrícolas e identificar riscos ambientais e operacionais.

Sua função é interpretar dados simulados da missão e gerar respostas técnicas, objetivas e úteis para tomada de decisão.

Você deve analisar variáveis como:

- temperatura;
- comunicação;
- bateria;
- oxigênio;
- estabilidade;
- luminosidade;
- vibração;
- risco ambiental.

Ao receber dados de um ciclo da missão, você deve:

1. Identificar o estado geral da missão;
2. Apontar os principais riscos;
3. Explicar o motivo dos alertas;
4. Recomendar ações operacionais;
5. Relacionar a resposta com sustentabilidade, eficiência energética e proteção agrícola.

Regras principais:

- Temperatura acima de 35 indica calor elevado.
- Temperatura acima de 40 indica risco crítico ambiental.
- Bateria abaixo de 50 indica atenção energética.
- Bateria abaixo de 20 indica risco crítico de energia.
- Comunicação abaixo de 60 indica comunicação instável.
- Comunicação abaixo de 30 indica falha crítica de comunicação.
- Oxigênio abaixo de 80 indica suporte operacional crítico.
- Estabilidade abaixo de 40 indica instabilidade operacional crítica.
- Vibração acima de 60 indica possível impacto ou falha mecânica.
- Risco ambiental acima de 70 indica área agrícola em risco alto.

Sua resposta deve seguir este formato:

STATUS DA MISSÃO:
[estável, atenção ou crítico]

PRINCIPAIS ALERTAS:
- alerta 1
- alerta 2
- alerta 3

ANÁLISE:
Explique de forma curta o que está acontecendo.

RECOMENDAÇÃO:
Indique ações práticas para a equipe da missão.

IMPACTO SUSTENTÁVEL:
Explique como a decisão ajuda a proteger recursos, reduzir desperdícios ou apoiar o agronegócio sustentável.
"""
```

```python
def gerar_alertas_logicos(ciclo):
    alertas = []

    if ciclo["temperatura"] > 40:
        alertas.append("Risco crítico ambiental por temperatura elevada")
    elif ciclo["temperatura"] > 35:
        alertas.append("Alerta de calor elevado")

    if ciclo["bateria"] < 20:
        alertas.append("Bateria crítica: ativar economia de energia")
    elif ciclo["bateria"] < 50:
        alertas.append("Atenção energética: bateria abaixo do ideal")

    if ciclo["comunicacao"] < 30:
        alertas.append("Falha crítica de comunicação")
    elif ciclo["comunicacao"] < 60:
        alertas.append("Comunicação instável")

    if ciclo["oxigenio"] < 80:
        alertas.append("Suporte operacional crítico")

    if ciclo["estabilidade"] < 40:
        alertas.append("Instabilidade operacional crítica")

    if ciclo["vibracao"] > 60:
        alertas.append("Possível impacto ou falha mecânica")

    if ciclo["risco_ambiental"] > 70:
        alertas.append("Área agrícola em risco ambiental alto")

    if not alertas:
        alertas.append("Nenhum alerta crítico identificado")

    return alertas
```

```python
def analisar_com_ia(ciclo):
    alertas_logicos = gerar_alertas_logicos(ciclo)

    prompt_usuario = f"""
Analise o seguinte ciclo da Mission TerraGuard:

Ciclo: {ciclo["ciclo"]}
Temperatura: {ciclo["temperatura"]} °C
Comunicação: {ciclo["comunicacao"]}%
Bateria: {ciclo["bateria"]}%
Oxigênio: {ciclo["oxigenio"]}%
Estabilidade: {ciclo["estabilidade"]}%
Luminosidade: {ciclo["luminosidade"]}
Vibração: {ciclo["vibracao"]}
Risco ambiental: {ciclo["risco_ambiental"]}

Alertas lógicos já detectados pelo sistema:
{alertas_logicos}

Gere uma análise operacional com recomendação para a missão.
"""

    resposta = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": prompt_usuario}
        ],
        temperature=0.4
    )

    return resposta.choices[0].message.content
```

```python
# Seleciona um ciclo específico para demonstrar a análise da IA
# Neste exemplo, usamos o ciclo 5 porque ele possui condições críticas na missão.

ciclo_analisado = dados_missao[4]

analise = analisar_com_ia(ciclo_analisado)

print("=" * 70)
print("ORBITAL AGROVISION — MISSION CONTROL AI")
print("=" * 70)
print(f"Ciclo analisado: {ciclo_analisado['ciclo']}")
print("-" * 70)

print("DADOS DO CICLO:")
print(f"Temperatura: {ciclo_analisado['temperatura']} °C")
print(f"Comunicação: {ciclo_analisado['comunicacao']}%")
print(f"Bateria: {ciclo_analisado['bateria']}%")
print(f"Oxigênio: {ciclo_analisado['oxigenio']}%")
print(f"Estabilidade: {ciclo_analisado['estabilidade']}%")
print(f"Luminosidade: {ciclo_analisado['luminosidade']}")
print(f"Vibração: {ciclo_analisado['vibracao']}")
print(f"Risco ambiental: {ciclo_analisado['risco_ambiental']}")
print("-" * 70)

print("ANÁLISE GERADA PELA IA:")
print(analise)
```

```python
for ciclo in dados_missao:
    print("=" * 70)
    print(f"ANÁLISE IA — CICLO {ciclo['ciclo']}")
    print("=" * 70)

    analise = analisar_com_ia(ciclo)
    print(analise)
    print("\n")
```
